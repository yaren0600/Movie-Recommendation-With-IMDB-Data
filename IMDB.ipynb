# Tabloların birleştirilmesi (merge)

users ve movies tablolarını ortak sütun olan iten_id üzerinden birleştiriyoruz. Daha sonra yeni oluşturulan data değişkenine yeni tabloyu kaydediyoruz.
Bu işlem SQL de bulunan join işlemidir.

```python
data = pd.merge(users, movies, on = "item_id") 
```

```python
data.tail()
```

```python
table = data.pivot_table(index = "user_id", columns = "title", values = "rating")
table.head()
```

```python
starwars = table ["Star Wars (1977)"]
starwars.head()
# Burada yapılan işlem, Star Wars (1977) filmi için tüm kullanıcıların verdiği puanları içeren bir seriyi (Series) tablodan çekmek.
```

```python
type(starwars)
```

```python
correlation = table.corrwith(starwars)
```

```python
correlation
```

```python
corr = pd.DataFrame(correlation, columns = ["Correlation"])
corr.dropna(inplace = True)
corr.head()
# Korelasyon değerini DataFrame'e dönüştürüp yeniden sıraladık.
```

```python
corr.sort_values("Correlation", ascending = False).head(15) 
# corr DataFrame'ini büyükten küçüğe sıralıyoruz(ascending = false)
# Bu işlem sayesinde bir filmle en benzer filmler bulunur.
```

```python
data.head()
```

```python
data.drop((["timestamp"]), axis = 1)
# Bu işlem sayesinde tabloda yer alan timestamp sütununu siliyoruz.(drop)
```

```python
ratings = pd.DataFrame(data.groupby("title")["rating"].mean())
# Rating değerinin ortalamasını aldık
ratings.sort_values("rating", ascending = False).head()
# Yeni oluşan ratings tablosunda yer alan rating değerleri büyükten küçüğe sıralanıp gösterilir.
```

```python
ratings["votes"] = pd.DataFrame(data.groupby("title")["rating"].count())
ratings.head()
# ratings tablosuna votes sütununu ekledik
```

```python
ratings.sort_values("votes", ascending = False).head()
```

```python
corr.head()
```

```python
corr = corr.join(ratings["votes"])
corr.head()
# Her filmin karşısında oy sayısı ve korelasyon değeri gösteriliyor. 
```

```python
corr[corr["votes"] > 100].sort_values("Correlation", ascending = False).head()
# 100 den fazla oy almış olan filmleri korelasyon değerlerini büyükten küçüğe sıralıyoruz.
# Bu filmlerden benzer olan 5 tanesi öneri olarak gösterilir.
```
